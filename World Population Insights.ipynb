```python
#importing necessary libraries for the project
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing data
data=pd.read_csv(r'C:\Users\User\Desktop\data_hub\world_population_data.csv')
```

```python
#Distribution of 2023 population in histogram
sns.histplot(data['2023 population'],bins=10,kde=True)
plt.title('Distribution of 2023 population')
plt.xlabel('Poplulation')
plt.ylabel('Frequency')
plt.show()
```

```python
#continents population in the year of 2023
data.groupby('continent')['2023 population'].mean().sort_values(ascending=False)
```

```python
#Population distribution in continentwise in 2023 according to bargraph
data.groupby('continent')['2023 population'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Population distribution in continentwise in 2023')
plt.xlabel('Continent')
plt.ylabel('Population Number')
plt.show()
```

```python
#Top 10  countries by population
top_10_countries=data.sort_values(by='2023 population',ascending=False).head(10)
```

```python
#Top 10  countries by population by bar graph
plt.figure(figsize=(8,6))
sns.barplot(x='country',y='2023 population',data=top_10_countries,palette='Accent')
plt.title('Top 10  countries by population')
plt.xlabel('Country Name')
plt.ylabel('2023 population')
plt.xticks(rotation=70)
plt.show()
```

```python
#Population and Density scatterplot
sns.scatterplot(x='2023 population',y='density (km²)',data=data)
plt.title('Population and Density scatterplot')
plt.xlabel('Population')
plt.ylabel('Density')
plt.show()
```

```python
#Population Density distribution across continents
data.groupby('continent')['area (km²)'].sum().sort_values(ascending=False)
```

```python
#Population Density distribution across continents according to violin plot
sns.violinplot(x='continent',y='density (km²)',data=data,palette='Dark2')
plt.title('Population Density distribution across continents')
plt.xlabel('Continent')
plt.ylabel('Population Density')
plt.show()
```

```python
#Population Density distribution across continents according to boxplot
sns.boxplot(x='continent',y='density (km²)',data=data,palette='PRGn',showfliers=False)
plt.title('Population Density distribution across continents')
plt.xlabel('Continent')
plt.ylabel('Population Density')
plt.show()
```

```python
top_10_over_populated_country=data.groupby('country',as_index=False)['density (km²)'].mean().sort_values(by='density (km²)',ascending=False).head(10)
```

```python
sns.barplot(x='country',y='density (km²)',data=top_10_over_populated_country,palette='Paired')
plt.xticks(rotation=70)
plt.show()
```

```python
#Population distribution in all continents in lineplot
data1.transpose().plot(figsize=(20,7))
plt.title('Population distribution in all continents')
plt.xlabel('Yearwise total Polulation')
plt.ylabel('Number of population')
plt.show()
```

```python
#Population Distribution in all continents according to boxplot
data1.transpose().boxplot(figsize=(20,7))
plt.title('Population Distribution in all continents')
plt.xlabel('Continents')
plt.ylabel('Total Population')
plt.show()
```

```python
#country wise population distribution in every year
data2=data.groupby('country')[['1970 population',
       '1980 population', '1990 population', '2000 population',
       '2010 population', '2015 population', '2020 population',
       '2022 population', '2023 population']].mean().sort_values(by='2023 population',ascending=False)
```

```python
#first 10 rows of the new dataset
data2.head(10)
```

```python
#Top 8 countries by population according to line chart
data2.head(8).transpose().plot(figsize=(12,7))
plt.title('Top 8 countries by population')
plt.xlabel('Year')
plt.ylabel('Average Population')
plt.xticks(rotation=60)
plt.show()
```

```python
#transforming growth rate into object datatype to float data type
data['growth rate_new']=data['growth rate'].replace('[^\d.]','',regex=True).astype(float)
```

```python
#transforming world percentage from object datatype to float data type
data['world_percentage_new']=data['world percentage'].replace('[^\d.]','',regex=True).astype(float)
```

```python
#Percentage of population in 6 continents according to pie chart
data.groupby('continent')['world_percentage_new'].sum().plot(kind='pie',autopct='%.1f%%')
plt.title('Percentage of population in 6 continents')
plt.ylabel('')
plt.show()
```

```python
#growth rate of population according to country wise
top_10_growth_rate=data.groupby('country',as_index=False)['growth rate_new'].sum().sort_values(by='growth rate_new',ascending=False).head(10)
top_10_growth_rate
```

```python
#Top 10 countries with highest growth rate
sns.barplot(x='country',y='growth rate_new',data=top_10_growth_rate,palette='Set1')
plt.title('Top 10 countries with highest growth rate')
plt.xlabel('Country')
plt.ylabel('Growth Rate')
plt.xticks(rotation=70)
plt.show()
```

```python
#Top 10 countries according to area coverage
large_country=data.groupby('country',as_index=False)['area (km²)'].sum().sort_values(by='area (km²)',ascending=False).head(10)
large_country
```

```python
#Top 10 Countries with area
sns.barplot(x='country',y='area (km²)',data=large_country,palette='Set2')
plt.title('Top 10 Countries with area')
plt.ylabel('Area')
plt.xticks(rotation=70)
plt.show()
```

```python
#Top 10 countries with highest density
data.groupby('country')['density (km²)'].mean().sort_values(ascending=False).head(10).plot(kind='barh')
plt.title('Top 10 countries with highest density')
plt.xlabel('Density')
plt.show()
```

```python
#Continents according to the density
ax=data.groupby('continent')['density (km²)'].sum().sort_values(ascending=False).plot(kind='barh')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Continents according to the density')
plt.xlabel('Density')
plt.show()
```

```python
#growth rate of the continents
growth_rate_by_continent=data.groupby('continent',as_index=False)['growth rate_new'].mean().sort_values(by='growth rate_new',ascending=False)
growth_rate_by_continent
```

```python
#Growth rate in continents
plt.figure(figsize=(6,4))
sns.barplot(x='continent',y='growth rate_new',data=growth_rate_by_continent,palette='brg')
plt.title('Growth rate in continents')
plt.ylabel('Average Growth Rate')
plt.xticks(rotation=60)
plt.show()
```

```python
#Growth rates distribution in 6 continents according to pie chart
data.groupby('continent')['growth rate_new'].sum().plot(kind='pie')
plt.title('Growth rates distribution in 6 continents')
plt.ylabel('')
plt.show()
```
